```python
# Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Get the data
data = pd.read_csv('/content/hot_dog_survey_spring25.xlsx - Form Responses 1 (1).csv')
data.head()
```

```python
# Get rid of columns we don't care about
results = []
letter = "y"

results.append(data.keys())

for index, row in data.iterrows():
  if letter.lower() in row['Are you currently enrolled as an undergraduate or graduate student at the University of Virginia?'].lower():
    results.append(row)


print(type(results))
```

```python
# Making the results into an array so easier to deal with
res = np.array(results)
print(res)
```

```python
# Deleting columns we don't care about
clean = np.delete(res, [0, 2, 3, 7, 8], 1)

# Renaming columns
clean[0][0] = "response"
clean[0][1] = "major"
clean[0][2] = "birthplace"
clean[0][3] = "year"
print(clean)
```

```python
# change yesses and nos
for i in range(1, len(clean)):
  if letter.lower() in clean[i][0]:
    clean[i][0] = "yes"
  else:
    clean[i][0] = "no"

print(clean)

# change virginia
```

```python
# Changing birthplaces
s = "Virginia"
s2 = "VA"
for i in range(1, len(clean)):
  if (s.lower() or s2.lower()) in str(clean[i][2]).lower():
    clean[i][2] = "virginia"
  else:
    clean[i][2] = "other"

print(clean)
```

```python
# Making plots
clean_df = pd.DataFrame(clean[1:], columns = clean[0])
#print(clean_df)
# Draw a nested barplot by species and sex

answer = clean_df['response']
bp = clean_df['birthplace']
yr = clean_df['year']
maj = clean_df['major']

from operator import le

cross1 = pd.crosstab(answer, bp)
cross2 = pd.crosstab(answer, yr)

boop = cross1.plot.bar(rot=0, stacked=True, color=sns.husl_palette(len(cross1.columns)))
plt.xlabel('Response')
legend_handles, _ = boop.get_legend_handles_labels()
boop.legend()

plt.ylabel('Count')
plt.show()

boop2 = cross2.plot.bar(rot=0, stacked=True, color=sns.husl_palette(len(cross2.columns)))
plt.xlabel('Response')
legend_handles, _ = boop2.get_legend_handles_labels()
boop2.legend()

plt.ylabel('Count')
plt.show()
```

```python
# Group and count responses by birthplace
yes_counts = sum((clean[1:, 0] == "yes") & (clean[1:, 2] == "virginia"))
no_counts = sum((clean[1:, 0] == "no") & (clean[1:, 2] == "virginia"))

other_yes_counts = sum((clean[1:, 0] == "yes") & (clean[1:, 2] == "other"))
other_no_counts = sum((clean[1:, 0] == "no") & (clean[1:, 2] == "other"))

# Bar chart
labels = ['Virginia', 'Other']
yes_vals = [yes_counts, other_yes_counts]
no_vals = [no_counts, other_no_counts]

x = np.arange(len(labels))
width = 0.35

plt.bar(x - width/2, yes_vals, width, label='Yes', color='skyblue')
plt.bar(x + width/2, no_vals, width, label='No', color='salmon')

plt.xticks(x, labels)
plt.title("Hotdog Opinion by Birthplace")
plt.xlabel("Birthplace")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
import pandas as pd

# Convert `clean` data to a DataFrame
df = pd.DataFrame(clean[1:], columns=clean[0])

# Contingency table between response and year
contingency_table = pd.crosstab(df['response'], df['year'])
print(contingency_table)


# Normalize by year to show proportions
proportion_table = contingency_table.div(contingency_table.sum(axis=0), axis=1)
print(proportion_table)


import seaborn as sns
import matplotlib.pyplot as plt

# Heatmap of proportions
plt.figure(figsize=(8, 5))
sns.heatmap(proportion_table, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title("Proportion of Yes/No Responses by Year")
plt.xlabel("Year")
plt.ylabel("Response")
plt.show()
```
